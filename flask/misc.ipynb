```python
import pandas as pd
import numpy as np
```

```python
def random_dates(start, end, n=100):
    start_u = start.value//10**9
    end_u = end.value//10**9
    
    return pd.to_datetime(np.random.randint(start_u, end_u, n), unit='s')
```

```python
start = pd.to_datetime('2009-01-01')
end = pd.to_datetime('2021-01-01')

da = random_dates(start, end)
```
